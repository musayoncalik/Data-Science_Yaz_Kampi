```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import svm, datasets
from sklearn.naive_bayes import GaussianNB
```

```python
# iris veri  setinin dahil edilmesi
iris = datasets.load_iris()
x = iris.data
y = iris.target
```

```python
x 
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size = 0.7, random_state = 42)
```

```python
X_train
```

```python
X_test
```

```python
y_test
```

```python
y_train
```

```python
# DT ile
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)
kesinlik = accuracy_score(y_test, dt_pred)
print(f"Karar Ağacı Modeli Kesinlik Scoru : {kesinlik}")
```

```python
#LR ile
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)
kesinlik = accuracy_score(y_test, lr_pred)
print(f"Logistic Regresyon Modeli Kesinlik Scoru : {kesinlik}")
```

```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")


ev_metrekaresi = np.array([50,100,150,200,250])
fiyat = np.array([10000,200000,300000,400000,500000])
```

```python
# modeli eğitme
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(ev_metrekaresi.reshape(-1,1), fiyat)
```

```python
metrekare = int(input("Evin metrekaresini giriniz : "))
tahmin_fiyat = model.predict([[metrekare]])
print(f"{metrekare} metrekarelik evin tahmini fiyatı {tahmin_fiyat[0]:.2f}")
```

```python
data = {
    'Hava Durumu': ['Güneşli', 'Güneşli', 'Bulutlu', 'Yağmurlu', 'Yağmurlu', 'Bulutlu', 'Güneşli', 'Yağışlı', 'Güneşli', 'Bulutlu' ],
    'Sıcaklık' : ['Sıcak', 'Sıcak', 'Ilık', 'Soğuk', 'Soğuk', 'Ilık', 'Sıcak', 'Soğuk', 'Sıcak', 'Ilık'],
    'Rüzgar' : ['Rüzgarsız', 'Rüzgarlı', 'Rüzgarlı', 'Rüzgarlı', 'Rüzgarsız', 'Rüzgarsız', 'Rüzgarlı', 'Rüzgarlı', 'Rüzgarsız', 'Rüzgarsız'],
    'Olumlu' : ['E', 'E', 'H', 'H', 'H','E', 'E', 'H', 'E', 'E']
} 
```

```python
df = pd.DataFrame(data)
df
```

```python
# Hava Durumu kolonu için : Güneşli : 0, Bulutlu : 1, Yağmurlu : 2
# Sıcaklık : Sıcak : 0, Ilık : 1, Soğuk : 2
# Rüzgar : Rüzgarsız : 0, Rüzgarlı : 1
# Olumlu : E : 1 , H : 0
```

```python
df['Hava Durumu'] = df['Hava Durumu'].map({'Güneşli': 0, 'Bulutlu': 1, 'Yağmurlu': 2, 'Yağışlı': 3})
df['Sıcaklık'] = df['Sıcaklık'].map({'Sıcak': 0, 'Ilık': 1, 'Soğuk': 2})
df['Rüzgar'] = df['Rüzgar'].map({'Rüzgarsız': 0, 'Rüzgarlı': 1})
df['Olumlu'] = df['Olumlu'].map({'E': 1, 'H': 0})
```

```python
df
```

```python
X = df.drop('Olumlu', axis = 1)
X
```

```python
y = df['Olumlu']
y
```

```python
model = DecisionTreeClassifier()
model.fit(X, y)
```

```python
hava_durumu = str(input('Hava Durumunu giriniz'))
sicaklik = str(input('Sıcaklık giriniz'))
ruzgar = str(input('Rüzgar durumunu giriniz'))
new_data = pd.DataFrame({f'Hava Durumu':[hava_durumu], 'Sıcaklık': [sicaklik], 'Rüzgar': [ruzgar]})

new_data['Hava Durumu'] = new_data['Hava Durumu'].map({'Güneşli': 0, 'Bulutlu': 1, 'Yağmurlu': 2, 'Yağışlı': 3})
new_data['Sıcaklık'] = new_data['Sıcaklık'].map({'Sıcak': 0, 'Ilık': 1, 'Soğuk': 2})
new_data['Rüzgar'] = new_data['Rüzgar'].map({'Rüzgarsız': 0, 'Rüzgarlı': 1})

#tahmin etme
tahmin = model.predict(new_data)

if tahmin[0] ==1:
    print("Bugün dışarı çıkabilirsiniz!")
else:
    print("Bugün Hava Şartları Çok uygun değil evinizde kalmanızı öneririm!")
    
```
