```python
# Veri görsellleştirme
# y = x^2

import matplotlib.pyplot as plt
plt.style.use('ggplot')

x = np.arange(1,11)  # [1 , 2, 3 .. 10]
y = x ** 2         # [1 , 4, 9, .. 100]
plt.title('Kare Fonksiyon')
plt.plot(x,y)
plt.show()
```

```python
# sinüs ve kosinüs grafiğini aynı yerde çiziniz.

x = np.linspace(0, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize = (8, 4))
plt.plot(x,y1, label = "Sin", color = "#79c2d0" )
plt.plot(x, y2, label = 'Cos', color = "#ff6f3c")
plt.title('Sin(x) ve Cos(x)')
plt.xlabel('X')
plt.legend()
plt.show()          
```

```python
dersler = ['Matematik', 'Fizik', 'Kimya ', 'Biyoloji']
notlar = [85,70, 55, 65]
renkler = ['#a2a8d3', '#f95959', '#ffc93c']

df = pd.DataFrame({'Ders': dersler, 'not': notlar})
df
```

```python
plt.pie(df['not'], labels=df['Ders'], colors=renkler, autopct = lambda p : f'%{p:.2f} \n {p * sum(df["not"])/100:.0f}')
plt.title("Ders Notları Dağılımı")
plt.show()
```

```python
plt.bar(df['Ders'], df['not'], color=renkler)
plt.title("Ders Notları")
plt.xlabel("Dersler")
plt.ylabel("Notlar")
bars = plt.bar(df['Ders'], df['not'], color=renkler)


# Her bar üzerine hem yüzdelik hem de değer yazdır
toplam = sum(df["not"])
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, 
             f'%{height/toplam*100:.2f}\n{height}', 
             ha='center', va='bottom')


plt.show()
```

```python
# Stok Takip Uygulaması

data = {
    'Ürün adı' : ["Elma", "Armut", "Ay çekirdeği", "Şeker", "Un", "Yağ", "Helva", "Nohut Dürümü", "Kayısı", "Fıstık", "Fındık", "Lokum"],
    'Stok' : np.random.randint(0,200,12),
    'Fiyat': np.random.randint(25,150,12),
    'Kategori': ['Meyve', 'Meyve', 'Çerez', 'Erzak', 'Erzak', 'Erzak', 'Atıştırmalık', 'Yemek', 'Meyve', 'Çerez', 'Çerez', 'Çerez']
}
```

```python
df1 = pd.DataFrame(data)
df1
```

```python
# Kdv siz verilmiştir. kdv li fiyatı da ayrı bir kolon ile gösterilmesi istenilmektedir.

df1['Kdv li fiyatı'] = df1['Fiyat'] * 1.2
df1
```

```python
# Kdv li fiyattan tüm ürünleri sattığımızı varsayalım. 
# Her ürün için ne kadar TL lik satış gelir elde ettiğimizi gösterelim

df1["Gelir"] = df1['Stok'] * df1["Kdv li fiyatı"]
df1
```

```python
# Kategoriye göre toplam gelirleri gruplayan listeyi veriniz.

gruplanmis_kategori = df1.groupby('Kategori')['Gelir'].sum()
gruplanmis_kategori
```

```python
# Gruplanmış ürünleri bar grafikte gösterin

plt.bar(gruplanmis_kategori.keys(), gruplanmis_kategori.values)
plt.title("Kategoriye göre toplam gelir")
plt.show()
```

```python
values = gruplanmis_kategori.values  # her kategoriye düşen ürün sayısı zaten sayılar

# Pie chart çizimi
plt.pie(values, 
        labels = gruplanmis_kategori.index, 
        autopct = lambda p: f'{p:.2f}%\n{int(p * sum(values)/100)}')

plt.show()
```
