```python
daf1 = pd.DataFrame({'id': [1,2,3,4,5], 'Name' : ['Ömer', 'Buse', 'Akça', 'SEna', 'Yusuf']})
daf1
```

```python
daf2 = pd.DataFrame({'id' : [1,2,3,4,5,6,7,8],
                    'Takım' :    ['Malatyaspor', 'Ümraniyespor', 'Galatasaray', 'Fenerbahçe', 'Beşiktaş', 'Manisaspor','Trabzonspor', 'Başakşehir']})
```

```python
daf2
```

```python
# inner join

pd.merge(daf1, daf2, on = 'id', how = 'inner')
```

```python
# outer join 

pd.merge(daf1, daf2, on='id', how='outer')
```

```python
# left join 

pd.merge(daf2, daf1 , on = 'id', how = 'left')
```

```python
# left join 

pd.merge(daf1, daf2 , on = 'id', how = 'left')
```

```python
# right join 

pd.merge(daf1, daf2 , on = 'id', how = 'right')
```
